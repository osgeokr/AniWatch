```python
# 모듈 호출
import tkinter as tk # Tkinter
from PIL import ImageTk, Image # Pillow
import cv2 as cv # OpenCV

# GUI 설계
win = tk.Tk() # 인스턴스 생성

win.title("AniWatch(야생동물 영상인식 딥러닝 소프트웨어) Ver.0.1") # 제목 표시줄 추가
win.geometry("920x640+50+50") # 지오메트리: 너비x높이+x좌표+y좌표
win.resizable(False, False) # x축, y축 크기 조정 비활성화

# 프레임 추가
frm = tk.Frame(win, width=720, height=480, padx=10, pady=10) # 프레임 너비, 높이 설정
frm.grid(row=0, column=0) # 프레임 행, 열 배치

# 프레임 라벨 추가
lblFrm = tk.Label(frm)
lblFrm.grid()

cap = cv.VideoCapture('otter.avi') # VideoCapture 객체 정의

def video_play():
    ret, frame = cap.read() # 프레임이 올바르게 읽히면 ret=True
    if not ret:
        cap.release() # 작업 완료 후 해제
        return
    frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    img = Image.fromarray(frame) # Image 객체로 변환
    imgtk = ImageTk.PhotoImage(image=img) # ImageTk 객체로 변환
    # OpenCV 동영상
    lblFrm.imgtk = imgtk
    lblFrm.configure(image=imgtk)
    lblFrm.after(42, video_play) # 프레임당 42밀리초 재생

video_play()
win.mainloop() #GUI 시작
```

```python
# 모듈 호출
import tkinter as tk # Tkinter
from PIL import ImageTk, Image # Pillow
import cv2 as cv # OpenCV
import os
from datetime import datetime
from tkinter import messagebox

# GUI 설계
win = tk.Tk() # 인스턴스 생성
win.title("AniWatch(야생동물 영상인식 딥러닝 소프트웨어) Ver.0.1") # 제목 표시줄 추가
win.geometry("920x640+50+50") # 지오메트리: 너비x높이+x좌표+y좌표
win.resizable(False, False) # x축, y축 크기 조정 비활성화

# 프레임 추가
frm = tk.Frame(win, width=720, height=480, padx=10, pady=10) # 프레임 너비, 높이 설정
frm.grid(row=0, column=0, columnspan=4) # 프레임 행, 열 배치

# 프레임 라벨 추가
lblFrm = tk.Label(frm)
lblFrm.grid()

# 동영상 파일 목록 수집
ext = [".avi", ".AVI"] # 파일 확장자
fileLst = [file for file in os.listdir("./") if file.endswith(tuple(ext))]

fileIdx = 0
cap = cv.VideoCapture(str(fileLst[fileIdx])) # VideoCapture 객체 정의

def show_frame():
    ret, frame = cap.read() # 프레임이 올바르게 읽히면 ret=True
    if not ret:
        cap.set(cv.CAP_PROP_POS_FRAMES, 0) # 프레임 위치 초기화
        return
    frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    img = Image.fromarray(frame) # Image 객체로 변환
    imgtk = ImageTk.PhotoImage(image=img) # ImageTk 객체로 변환
    # OpenCV 동영상
    lblFrm.imgtk = imgtk
    lblFrm.configure(image=imgtk)
    lblFrm.after(5, show_frame) # 프레임당 5밀리초 재생

# 날짜시간 라벨 추가
lblTime = tk.Label(win, text="Date Created(생성날짜): "+
                   str(datetime.fromtimestamp(os.stat(fileLst[fileIdx]).st_ctime)), padx=10) # 날짜시간
lblTime.grid(row=1, column=0, sticky="w")

def click_btnPlay():
    # messagebox.showinfo(title="GUI 시험", message="클릭했습니다")
    global cap, fileIdx
    lblTime.configure(text="Date Created(생성날짜): "+ str(datetime.fromtimestamp(os.stat(fileLst[fileIdx]).st_ctime)))
    cap = cv.VideoCapture(str(fileLst[fileIdx])) # VideoCapture 객체 정의
    show_frame()

def click_btnPrev():
    global cap, fileIdx
    fileIdx = fileIdx - 1 if (fileIdx !=0) else 0
    lblTime.configure(text="Date Created(생성날짜): "+ str(datetime.fromtimestamp(os.stat(fileLst[fileIdx]).st_ctime)))
    cap = cv.VideoCapture(str(fileLst[fileIdx])) # VideoCapture 객체 정의
    show_frame()

def click_btnNext():
    global cap, fileIdx
    lblTime.configure(text="Date Created(생성날짜): "+ str(datetime.fromtimestamp(os.stat(fileLst[fileIdx]).st_ctime)))
    fileIdx = fileIdx + 1 if (fileIdx < len(fileLst)-1) else len(fileLst)-1
    cap = cv.VideoCapture(str(fileLst[fileIdx])) # VideoCapture 객체 정의
    show_frame()
    
# 재생 버튼 추가
btnPlay = tk.Button(win, width=15, text="Play(재생)", command=click_btnPlay)
btnPlay.grid(row=1, column=1)
    
# 이전 버튼 추가
btnPrev = tk.Button(win, width=15, text="Prev(이전)", command=click_btnPrev)
btnPrev.grid(row=1, column=2)

# 다음 버튼 추가
btnNext = tk.Button(win, width=15, text="Next(다음)", command=click_btnNext)
btnNext.grid(row=1, column=3)

# 종 식별 버튼 추가
rbtnGroup = tk.LabelFrame(win, text = "Species ID(종 식별)")
rbtnGroup.grid(row=0, column=4)
species = tk.IntVar()
tk.Radiobutton(rbtnGroup, text="Water deer(고라니)", variable=species, value=0).grid(sticky="w")
tk.Radiobutton(rbtnGroup, text="Racoon dog(너구리)", variable=species, value=1).grid(sticky="w")
tk.Radiobutton(rbtnGroup, text="Roe deer(노루)", variable=species, value=2).grid(sticky="w")
tk.Radiobutton(rbtnGroup, text="Marten(담비)", variable=species, value=3).grid(sticky="w")
tk.Radiobutton(rbtnGroup, text="Wild boar(멧돼지)", variable=species, value=4).grid(sticky="w")
tk.Radiobutton(rbtnGroup, text="Leopard cat(삵)", variable=species, value=5).grid(sticky="w")
tk.Radiobutton(rbtnGroup, text="Badger(오소리)", variable=species, value=6).grid(sticky="w")
tk.Radiobutton(rbtnGroup, text="Not listed(목록 없음)", variable=species, value=7).grid(sticky="w")

show_frame()
win.mainloop() #GUI 시작
```
